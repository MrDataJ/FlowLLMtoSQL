```python
# Dependencies
from flipside import Flipside 
import requests
import os
import json
import time
from dotenv import load_dotenv
import openai
```

```python
# My first query
load_dotenv()
FLIPSIDE_API_KEY = os.getenv('FLIPSIDE_API_KEY')

# Initialize `Flipside` with your API Key and API Url
flipside = Flipside(FLIPSIDE_API_KEY, "https://api-v2.flipsidecrypto.xyz")

sql = """
SELECT 
  date_trunc('hour', block_timestamp) as hour,
  count(distinct tx_hash) as tx_count
FROM ethereum.core.fact_transactions 
WHERE block_timestamp >= GETDATE() - interval'7 days'
GROUP BY 1
"""

# Run the query against Flipside's query engine and await the results
query_result_set = flipside.query(sql)
print(query_result_set.json())
```

```python
# List of each table
flow_tbl_list = [
    'flow.core.dim_allday_metadata', 
    'flow.core.dim_contract_labels', 
    'flow.core.dim_moment_metadata', 
    'flow.core.dim_swap_pool_labels', 
    'flow.core.dim_topshot_metadata', 
    'flow.core.dim_validator_labels', 
    'flow.core.ez_bridge_transactions', 
    'flow.core.ez_nft_sales', 
    'flow.core.ez_staking_actions', 
    'flow.core.ez_swaps', 
    'flow.core.ez_token_transfers', 
    'flow.core.fact_blocks', 
    'flow.core.fact_events', 
    'flow.core.fact_hourly_prices', 
    'flow.core.fact_transactions'
    ]

flow_tbl_schema = {}

# Loop through each table and get the schema
for flow_tbl in flow_tbl_list:
    sql = f"""
      SHOW COLUMNS IN {flow_tbl}
    """
    query_result_set = flipside.query(sql)
    flow_tbl_schema[flow_tbl] = query_result_set

print(flow_tbl_schema)
```

```python
# Sample each table for the first 3 rows
flow_tbl_sample = {}

# Loop through each table and get the schema
for flow_tbl in flow_tbl_list:
    sql = f"""
      SELECT * FROM {flow_tbl}
      LIMIT 3
    """
    query_result_set = flipside.query(sql)
    flow_tbl_sample[flow_tbl] = query_result_set

print(flow_tbl_sample)
```

```python
schmea = flow_tbl_schema
sample = flow_tbl_sample
```

load_dotenv()
openai.api_key = os.getenv("OPENAI_API")

def ask_openai(message):
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are an SQL master. Only you can answer the following questions with strict SQL code, nothing else."},
            {"role": "user", "content": message},
        ]
    )
    
    answer = response.choices[0].message.content.strip()
    return answer
